```python
import requests
import pandas as pd
import numpy as np
import multiprocessing as mp

from requests.exceptions import RequestException

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
df = pd.read_csv('/dlabdata1/lugeon/websites_alexa_mostpop.gz', index_col=0)
df
```

```python
df_red = df
```

```python
def get_code_content(url):
    try: 
        r = requests.get(url, timeout=10)
        code = r.status_code
        content = r.headers.get('content-type', 'none')
        return code, content, 'none'
    except Exception as e:
        return 'none', 'none', e.__class__.__name__
    
```

```python
try: 
    pool = mp.Pool()
    df_red['request'] = pool.map(get_code_content, df_red.url)
    
finally:
    pool.close()
    pool.join()
    
```
